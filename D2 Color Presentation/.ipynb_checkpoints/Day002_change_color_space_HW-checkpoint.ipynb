## Goal:  

present Lena in 'HLS' and 'LAB' color space

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
```

# NOTE! OpenCV: BGR Manner; Matplotlib: RGB Manner

[matplotlib](https://blog.csdn.net/saltriver/article/details/78173219)

```python
img_path = 'C:/Users/CSK/CSK-DL-CV-Marathon/images/lena.png'
img = cv2.imread(img_path, cv2.IMREAD_COLOR)

if img is None:
    sys.exit("could not read image")

# OpenCV color convertion
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
img_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    
# for matplotlib to show (convert bgr to rgb)
img_plt = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_plt_hls = cv2.cvtColor(img_hls, cv2.COLOR_BGR2RGB)
img_plt_lab = cv2.cvtColor(img_lab, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(1,3,1)
plt.imshow(img_plt)
ax.set_title('RGB')
ax = fig.add_subplot(1,3,2)
plt.imshow(img_plt_hls)
ax.set_title('HLS')
ax = fig.add_subplot(1,3,3)
plt.imshow(img_plt_lab)
ax.set_title('LAB')
```

```python
# OpenCV show

# concatenate images
colorSpace_concat = np.hstack((img, img_hls, img_lab))

while True:
    k = cv2.waitKey(0)
    cv2.imshow('lena bgr-hls-lab', colorSpace_concat)
    
    # press 's or S' to save images
    if k == ord("s") or k == ord("S"):
        cv2.imwrite('lena_bgr_hls_lab.png', colorSpace_concat)
        print("image saved")
        
    if k == 27:
        cv2.destroyAllWindows()
        break
    
```
