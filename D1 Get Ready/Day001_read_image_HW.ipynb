```python
import cv2 as cv
import numpy as np
import sys

img_path = 'C:/Users/CSK/CSK-DL-CV-Marathon/images/lena.png'

# load image and convert it into color one
img = cv.imread(img_path, cv.IMREAD_COLOR)

# check
if img is None:
    sys.exit("Could not read the image.")
    
# method 1: (split func)
b1, g1, r1 = cv.split(img)
# create empty channel
zeros = np.zeros(b1.shape, np.uint8)
# merge 3 channels
b1 = cv.merge((b1, zeros, zeros))
g1 = cv.merge((zeros, g1, zeros))
r1 = cv.merge((zeros, zeros, r1))

# method 2: (make a copy and set other channels to zeros)
b2 = img.copy()
g2 = img.copy()
r2 = img.copy()
# set g,r channels to 0
b2[:, :, 1] = 0
b2[:, :, 2] = 0
# set b,r channels to 0
g2[:, :, 0] = 0
g2[:, :, 2] = 0
# set b,g channels to 0
r2[:, :, 0] = 0
r2[:, :, 1] = 0

# concatenate images
bgr_concat = np.hstack((b1, g1, r1))

# keep showing image
while True:

    k = cv.waitKey(0)
    
    # m1 results
    cv.imshow('blue 1', b1)
    cv.imshow('green 1', g1)
    cv.imshow('red 1', r1)
    
    # m2 results
    cv.imshow('blue 2', b2)
    cv.imshow('green 2', g2)
    cv.imshow('red 2', r2)
    
    cv.imshow('Lena_bgr', bgr_concat)
       
    # press 's or S' to save images
    if k == ord("s") or k == ord("S"):
        cv.imwrite('Lena_b.png', b1)
        cv.imwrite('Lena_g.png', g1)
        cv.imwrite('Lena_r.png', r1)
        cv.imwrite('Lena_bgr.png', bgr_concat)
        print("image saved")
    
    # press ESC to close the display windows and break the while loop
    if k == 27:
        cv.destroyAllWindows()
        break
```

## Image histogram:

```python
import matplotlib.pyplot as plt

color = ['b', 'g', 'r']
for channel, col in enumerate(color):
    histr = cv.calcHist([img], [channel], None, [256], [0,256])
    plt.plot(histr, color = col)
    plt.xlim([0,256])

plt.title('histogram for BGR channel')
plt.show()
```
